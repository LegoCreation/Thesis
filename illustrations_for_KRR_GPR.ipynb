```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.kernel_ridge import KernelRidge

# Generate some data
np.random.seed(0)
X = np.sort(5 * np.random.rand(80, 1), axis=0)
y = np.sin(X).ravel()
y += 0.5 * np.random.randn(80)

# Define a linear kernel function
def linear_kernel(x1, x2):
    return np.dot(x1, x2.T)

# Define a polynomial kernel function
def polynomial_kernel(x1, x2, d):
    return (1 + np.dot(x1, x2.T)) ** d

# Create a range of alpha values to test
alphas = np.logspace(-5, 2, 200)

# Create a range of degree values for the polynomial kernel
degrees = np.arange(1, 11)

# Initialize arrays to store results
mse_linear = np.zeros_like(alphas)
mse_polynomial = np.zeros((len(alphas), len(degrees)))

# Perform kernel ridge regression with linear kernel
for i, alpha in enumerate(alphas):
    kr = KernelRidge(kernel=linear_kernel, alpha=alpha)
    kr.fit(X, y)
    y_pred = kr.predict(X)
    mse_linear[i] = np.mean((y_pred - y) ** 2)

# Perform kernel ridge regression with polynomial kernel
for i, alpha in enumerate(alphas):
    for j, degree in enumerate(degrees):
        kr = KernelRidge(kernel=lambda x1, x2: polynomial_kernel(x1, x2, degree), alpha=alpha)
        kr.fit(X, y)
        y_pred = kr.predict(X)
        mse_polynomial[i, j] = np.mean((y_pred - y) ** 2)

```

```python
fig, axs = plt.subplots(1, 1, figsize=(5, 5))

# Plot MSE vs. alpha for the polynomial kernel
for j, degree in enumerate(degrees):
    axs.plot(alphas, mse_polynomial[:, j], label=f'Degree {degree}')
axs.set_xscale('log')
axs.set_xlabel('Alpha')
axs.set_ylabel('MSE')
axs.set_title('Kernel Ridge Regression with Polynomial Kernel')
axs.legend()

fig.savefig("./img/illustration_KRR_MSE.png", dpi = 100)
# Add scatter plot of original data and predicted values for all polynomial degrees
fig, axs = plt.subplots(2, 5, figsize=(15, 8))

for j, degree in enumerate(degrees):
    row = j // 5
    col = j % 5

    # Plot original data
    axs[row, col].scatter(X, y, c='k', label='Data')

    # Plot predicted values for polynomial kernel
    alpha = 0.01
    kr = KernelRidge(kernel=lambda x1, x2: polynomial_kernel(x1, x2, degree), alpha=alpha)
    kr.fit(X, y)
    y_pred = kr.predict(X)
    axs[row, col].plot(X, y_pred, c='r', label=f'Degree {degree}')
    axs[row, col].set_title("degree d = " + str(degree))
fig.suptitle("Scatter plot of KRR using polynomial kernels")
fig.savefig("./img/illustration_KRR_scatter.png", dpi = 100)
```

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Plot original data
axs[0].scatter(X, y, c='k', label='Data')

# Plot predicted values for linear kernel
alpha = 0.01
kr = KernelRidge(kernel=linear_kernel, alpha=alpha)
kr.fit(X, y)
y_pred = kr.predict(X)
axs[0].plot(X, y_pred, c='r', label='Linear kernel')
axs[0].set_xlabel('X')
axs[0].set_ylabel('y')
axs[0].set_title(f'Kernel Ridge Regression with Linear Kernel, alpha={alpha:.2f}')
axs[0].legend()
```
